```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm
from matplotlib.ticker import LinearLocator

# Make data.
U = np.arange(-5, 5, 0.25)
V = np.arange(-5, 5, 0.25)
U, V = np.meshgrid(U, V)

r = 2
Lambda = 0.5


plt.figure(figsize=(12, 4))
#for tau=0
tau = 0
plt.subplot(1, 3, 1)
Pu =  (Lambda/2)*np.exp(-(r-U*V)**2/2) * np.exp(-0.5 * tau * U * U) * np.exp(-0.5 * tau * V * V)
plt.title('tau=0')
plt.contourf(U, V, Pu)
plt.axis('Scaled')

#for tau=0.5
tau = 0.5
plt.subplot(1, 3, 2)
Pu =  (Lambda/2)*np.exp(-(r-U*V)**2/2) * np.exp(-0.5 * tau * U * U) * np.exp(-0.5 * tau * V * V)
plt.title('tau=0.5')
plt.contourf(U, V, Pu)
plt.axis('Scaled')

#for tau=1
tau = 1
plt.subplot(1, 3, 3)
Pu =  (Lambda/2)*np.exp(-(r-U*V)**2/2) * np.exp(-0.5 * tau * U * U) * np.exp(-0.5 * tau * V * V)
plt.title('tau =1')
plt.contourf(U, V, Pu)
plt.axis('Scaled')


plt.show()
```
